**Overview of Naive Bayes Classification**

**Naive Bayes** is a probabilistic classification algorithm based on **Bayes’ Theorem**. It assumes that the features used for prediction are **independent of each other given the class label**. Because of this assumption, it is called **“naive.”**

Naive Bayes is widely used for **text classification, spam detection, sentiment analysis, and medical diagnosis**.

---

**Bayes Theorem**

P(C | X) = ( P(X | C) * P(C) ) / P(X)

Where:

* P(C | X) → Posterior probability of class C given features X
* P(X | C) → Likelihood of features X given class C
* P(C) → Prior probability of class C
* P(X) → Evidence probability of features X

---

**Naive Bayes Classification Formula**

For multiple features:

P(C | X1, X2, ..., Xn) ∝ P(C) * P(X1 | C) * P(X2 | C) * ... * P(Xn | C)

The predicted class is:

C* = argmax_C [ P(C) * Π P(Xi | C) ]

---

**Key Terms**

**1. Prior Probability**

Probability of a class before observing the data.

P(C)

---

**2. Likelihood**

Probability of observing the feature given the class.

P(X | C)

---

**3. Posterior Probability**

Updated probability of the class after observing the features.

P(C | X)

---

**4. Evidence**

Total probability of observing the data.

P(X)

---

**5. Independence Assumption**

Naive Bayes assumes all features are **conditionally independent given the class**.

P(X1, X2, ..., Xn | C) = P(X1 | C) * P(X2 | C) * ... * P(Xn | C)

---

**Classification Rule**

The class with the **highest posterior probability** is selected as the predicted class.


**Dataset Loading and Inspection**

```python
import pandas as pd

iris_data = pd.read_csv('drive/MyDrive/Datasets For ML/Iris.csv')

print("Columns: ", iris_data.columns)
print("\nShape: ", iris_data.shape)
print("\nDuplicated: ", iris_data.duplicated().sum())
print("\nNull Values:", iris_data.isnull().sum().sum())
print("\nFirst 2 rows:")
iris_data.head(2)
```

**Data Cleaning**

```python
iris_data = iris_data.drop('Id', axis=1)
iris_data.shape
```

**Feature–Label Separation**

```python
X, Y = iris_data.drop('Species', axis=1), iris_data['Species']
```

**Train–Test Split (Stratified)**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.25, stratify=Y, random_state=5
)

print("Train Shape:", X_train.shape)
print("Test Shape:", X_test.shape)
```

**Why GaussianNB for Iris?**

Features are continuous

Data approximately follows Gaussian distribution

Small dataset → parametric model works well

**Gaussian Naive Bayes Model**

```python
from sklearn.naive_bayes import GaussianNB

gnb_iris = GaussianNB()
gnb_iris.fit(X_train, Y_train)
```

```python
gnb_iris.get_params()
```

**Model Predictions and Probabilities**

```python
Y_pred_iris = gnb_iris.predict(X_test)
print("Predicted Labels:\n", Y_pred_iris[:10])

print("\nPredicted Probabilities:\n", gnb_iris.predict_proba(X_test)[:10])
```

```python
gnb_iris.score(X_test, Y_test)
```

**Model Evaluation**

```python
from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(Y_test, Y_pred_iris))
print("Classification Report:\n", metrics.classification_report(Y_test, Y_pred_iris))
print("Confusion Matrix:\n", metrics.confusion_matrix(Y_test, Y_pred_iris))
```

**Conclusion**

The **Gaussian Naive Bayes** model performed best on the Iris dataset. It achieved high accuracy because the dataset contains continuous and well-separated features. The model required minimal parameter tuning while still providing effective classification performance.
