**NumPy Library**

NumPy is a Python library used for working with numbers and numerical data. It provides support for arrays and matrices and allows us to perform mathematical operations on large amounts of data easily and efficiently. NumPy is mainly used for calculations, data storage in array form, and scientific computing in Machine Learning.

**Importing NumPy**

NumPy is imported to perform numerical operations using arrays.

```python
import numpy as np
```

**1D Array in NumPy**

**Creating a 1D Array:** A 1D array stores multiple values in a single line.

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr)
arr
```

**Accessing Elements:** Elements are accessed using index positions starting from 0.

```python
print(arr[0])   # First element
print(arr[3])   # Fourth element
```

**Array Length:** Find the number of elements in the array.

```python
print(len(arr))
```

**Basic Operations on 1D Array**


Arithmetic operations are applied to all elements at once.

```python
print(arr + 5)
print(arr - 2)
print(arr * 2)
print(arr / 2)
```

**Mathematical Functions:** NumPy provides built-in mathematical functions.

```python
print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
```

**Changing Array Values:** Modify elements using index positions.

```python
arr[2] = 100
print(arr)
```

**Array Data Type:** Check the type of elements stored in the array.

```python
print(arr.dtype)
```

**Properties of 1D NumPy Array**


**Size of Array:** Size gives the total number of elements in the array.

```python
print(arr.size)
```

**Number of Dimensions:** It shows how many dimensions the array has.

```python
print(arr.ndim)
```

**Shape of Array:** Shape shows the structure of the array.

```python
print(arr.shape)
```

**Slicing in 1D Array**

**Basic Slicing:** Slicing is used to extract a part of the array.

```python
print(arr[1:4])   # From index 1 to 3
```

**Slicing from Beginning:** If starting index is not given, slicing starts from 0.

```python
print(arr[:3])
```

**Slicing till End:** If ending index is not given, slicing goes till last element.

```python
print(arr[2:])
```

**Negative Indexing:** Negative index accesses elements from the end.

```python
print(arr[-1])
print(arr[-3:])
```

**Iterating Through 1D Array**

**Using for Loop:** Looping through each element.

```python
for i in arr:
    print(i)
```

**Copying a 1D Array**

**Copy Method:** Creates a separate copy of the array.

```python
arr_copy = arr.copy()
arr_copy[0] = 999
print(arr)
print(arr_copy)
```

**Reshaping 1D Array**

**Reshape to 2D:** Convert 1D array into 2D form.

```python
new_arr = arr.reshape(1, 5)
print(new_arr)
```

**Checking Array Type**

**Type of Variable:** Confirms it is a NumPy array.



```python
print(type(arr))
```

**2D Arrays in NumPy**

**Creating a 2D Array:** A 2D array stores data in rows and columns.

```python
arr2d = np.array([[1, 2, 3],
                  [4, 5, 6]])
print(arr2d)
```

**Accessing Elements in 2D Array:** Access elements using row and column index.

```python
print(arr2d[0, 1])   # Element in 1st row, 2nd column
```

**Size of 2D Array**

```python
print(arr2d.size)
```

**Number of Dimensions**

```python
print(arr2d.ndim)
```

**Shape of 2D Array**

```python
print(arr2d.shape)
```

**Slicing in 2D Array**

```python
print(arr2d[:, 1])    # All rows, second column
print(arr2d[0, :])    # First row
```

**Iterating Through 2D Array**

```python
for row in arr2d:
    for element in row:
        print(element)
```

**NumPy Array Creation Functions**

**Zeros Array:** Creates an array filled with zeros.

```python
z = np.zeros(5)
print(z)
```

**Ones Array:** Creates an array filled with ones.

```python
o = np.ones(4)
print(o)
```

**Arange Function:** Creates values within a range.

```python
a = np.arange(1, 10, 2)
print(a)
```

**Linspace Function:** Creates evenly spaced values.

```python
l = np.linspace(1, 10, 5)
print(l)
```

**Comparison and Boolean Operations**

**Comparison Operators:** Used to compare array values.

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr > 25)
print(arr == 20)
```

**Boolean Indexing:**Extract elements based on condition.

```python
print(arr[arr > 25])
```

**Logical Operations**

```python
print(np.logical_and(arr > 15, arr < 45))
print(np.logical_or(arr < 15, arr > 45))
```

**Random Numbers in NumPy**

**Random Integer Values**

```python
r1 = np.random.randint(1, 100, 5)
print(r1)
```

**Random Float Values**

```python
r2 = np.random.rand(5)
print(r2)
```

**Random 2D Array**

```python
r3 = np.random.rand(2, 3)
print(r3)
```

**Setting Seed Value:** Ensures same random output every time.

```python
np.random.seed(10)
print(np.random.rand(3))
```

**Shuffle:** Shuffle randomly changes the order of elements in the same array.

```python
arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)
print(arr)
```

**Choice:** Choice selects random elements from an array.

```python
arr = np.array([10, 20, 30, 40, 50])
c = np.random.choice(arr)
print(c)
```

**Choice with Multiple Values:** Select more than one random value.

```python
c2 = np.random.choice(arr, size=3)
print(c2)
```

**Choice with Replacement:** Allows repeated values.

```python
c3 = np.random.choice(arr, size=5, replace=True)
print(c3)
```

**Choice without Replacement:** Each value is selected only once.

```python
c4 = np.random.choice(arr, size=5, replace=False)
print(c4)
```

**Permutation:** Returns a new shuffled array without changing original.

```python
arr = np.array([1, 2, 3, 4, 5])
p = np.random.permutation(arr)
print(p)
print(arr)
```

**Random Sample:** Generates random float values between 0 and 1.

```python
s = np.random.random(4)
print(s)
```

**Random Normal Distribution:** Generates values following normal distribution.

```python
n = np.random.normal(0, 1, 5)
print(n)
```

**Random Uniform Distribution:** Generates values within a given range.

```python
u = np.random.uniform(10, 20, 5)
print(u)
```
