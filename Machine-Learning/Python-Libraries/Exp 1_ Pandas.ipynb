**Pandas Library**

Pandas is a Python library used for data analysis and data manipulation.
It helps in reading datasets, cleaning data, handling missing values, and preparing data for Machine Learning models.
Pandas mainly works with tables of data called DataFrames.

**Important Pandas Data Structures**

**Series:** A one-dimensional labeled array, like a single column of a table.

**DataFrame:** A two-dimensional table consisting of rows and columns.
Most ML datasets are handled using DataFrames.

**Importing Pandas**

```python
import pandas as pd
```

**Handling Missing Values (Demonstration Purpose)**

**Note:** The Iris dataset does not contain missing values.
To understand missing value handling, some values are intentionally replaced with NaN.

**Creating Missing Values Artificially**

```python
iris_missing = iris_data.copy()

iris_missing.iloc[0, 0] = np.nan
iris_missing.iloc[4, 2] = np.nan
iris_missing.iloc[147, 3] = np.nan

iris_missing.head()
```

**Handling Missing Values (Very Important in ML)**

**Checking Missing Values:** Counts missing values in each column.

```python
iris_missing.isnull().sum()
```

**Dropping Missing Values:** Removes rows containing missing data.

```python
iris_dropped = iris_missing.dropna()
iris_dropped.head()
```

**Filling Missing Values:** Replaces missing values with 0.

```python
iris_filled = iris_missing.fillna(0)
iris_filled.head()
```

**Filling Missing Values with Mean (Better for ML)**

```python
iris_mean_filled = iris_missing.fillna(iris_missing.mean(numeric_only=True))
iris_mean_filled.head()
```

**Checking Correlation:** Shows relationship between numeriacl features.

*   Values close to -1 indicate strong negative relationship

*   Values near 0 indicate weak or no relationship

*   Petal length and petal width are highly correlated

*   Sepal length has moderate correlation with petal features

*   Sepal width shows weak correlation


**Why Correlation Matters in ML**

*   Removes redundant features

*   Simplifies model training

*   Improves model performance

*   Reduces overfitting

**Checking Unique Values:** Displays all distinct categories present in a column.

Used for grouped statistical analysis.
