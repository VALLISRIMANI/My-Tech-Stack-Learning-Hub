**Matplotlib Library**

Matplotlib is the core Python library for data visualization. It converts numerical data into visual form for analysis, comparison, and presentation. In ML it is used for
Understanding data distribution,Detecting patterns and outliers, Visualizing relationships between features, Monitoring model performance, feature behavior analysis, and result communication.

Matplotlib works directly with NumPy arrays, Pandas Series/DataFrames, and matrices.

*    pyplot provides a MATLAB-like interface.

*    Handles figures, axes, plots, labels, titles.

**Importing Matplotlib:** It imports the pyplot submodule of matplotlib and gives it the alias plt for easier use in plotting.

```python
import matplotlib.pyplot as plt
```

**Basic Plotting (Function Visualization):** Defines a mathematical function.

```python
def equation(x):
    return x**4 + 3*x**3 + 3*x**2 - 2*x + 1
```

X: input values

Y: computed outputs

Used to plot f(x)

```python
X = range(-5, 5)
Y = [equation(x) for x in X]
```

**Rule to Remember**

**Every plotting section must follow this order:**

1. plt.figure()

2. plotting commands

3. labels, title, legend

4. plt.tight_layout() if multiple plots

5. plt.show()

6. plt.close() or plt.clf()

If you follow this, plots will never mix.

**Subplots (Multiple Plots in One Figure)**

Divides figure into grid (rows, cols, index)

Useful for comparing variations

**plt.subplot(rows, columns, index)**

rows: total number of rows in the figure

columns: total number of columns in the figure

index: position of the subplot, counted left to right, top to bottom, starting from 1


**Used heavily in:**

Feature comparison

Model performance plots

```python
plt.subplot(2, 2, 1) # 2 rows, 2 columns, selects the 1st subplot (top-left)
plt.subplot(3, 2, 2) # 3 rows, 2 columns, selects the 2nd subplot (top-right)
```

**Important note:** All subplots in one figure should use the same rows and columns.
Mixing 2×2 and 3×2 layouts in the same figure will overwrite or distort plots.

```python
plt.subplot(2, 2, 1)
plt.plot(X, Y, marker="*", color="blue", linestyle="-")
plt.xlabel("X")
plt.ylabel("f(X)")

plt.subplot(2, 2, 2)
plt.plot(X, Y, marker="*", color="green", linestyle="--")
plt.xlabel("X")
plt.ylabel("f(X)")

plt.subplot(2, 2, 3)
plt.plot(X, Y, marker="*", color="orange", linestyle="-.")
plt.xlabel("X")
plt.ylabel("f(X)")

plt.subplot(2, 2, 4)
plt.plot(X , Y, marker="*", color="red", linestyle=":")
plt.xlabel("X")
plt.ylabel("f(X)")

plt.show()
```

**Check: The above plots look mixed or cluttered**

**Correct Way to Create Subplots (FIX)**

Always start with a new figure

This isolates the subplot grid.

```python
plt.figure(figsize=(10, 5))   # creates a fresh figure

plt.subplot(2, 2, 1)
plt.plot(X, Y, marker="*", color="blue", linestyle="-")
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Blue -")

plt.subplot(2, 2, 2)
plt.plot(X, Y, marker="*", color="green", linestyle="--")
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Green --")

plt.subplot(2, 2, 3)
plt.plot(X, Y, marker="*", color="orange", linestyle="-.")
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Orange -.")

plt.subplot(2, 2, 4)
plt.plot(X, Y, marker="*", color="red", linestyle=":")
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Red :")

plt.tight_layout()   # prevents overlap
plt.show()
```

**Mandatory Best Practice for Next Cells:** After every major plot block, clear the figure.

```python
plt.clf() # or plt.close() - Use plt.close() when moving to a completely different plot type.
```

**Scatter Plot:** A scatter plot is a graph that shows the relationship between two numerical features.
Each point represents a single observation.

**Uses in ML:**

Identify correlations between features.

Detect outliers.

Visualize trends.

```python
import pandas as pd
import numpy as np

file_path = '/content/drive/My Drive/Datasets For ML/Iris.csv'
```

```python
iris = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(iris["SepalLengthCm"], iris["SepalWidthCm"], color="lightgreen")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Feature Relationship")
plt.show()

plt.close()
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(iris["SepalLengthCm"], iris["PetalLengthCm"], color="steelblue")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Feature Relationship")
plt.grid(True)
plt.show()

plt.close()
```

**Class-wise Scatter Plot (ML Standard)**

A class-wise scatter plot shows data points grouped by their class or category.
Different classes are displayed using colors or markers.

**Uses in ML:**

Visualize separation between classes.

Understand class distribution.

Useful in classification problems.

```python
plt.figure(figsize=(10, 4))

plt.scatter(
    iris["PetalLengthCm"][0:50],
    iris["PetalWidthCm"][0:50],
    color="darkred",
    label="Setosa"
)

plt.scatter(
    iris["PetalLengthCm"][50:100],
    iris["PetalWidthCm"][50:100],
    color="navy",
    label="Versicolor"
)

plt.scatter(
    iris["PetalLengthCm"][100:150],
    iris["PetalWidthCm"][100:150],
    color="forestgreen",
    label="Virginica"
)

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Class Separation")
plt.legend()
plt.grid(True)
plt.show()

plt.close()
```
