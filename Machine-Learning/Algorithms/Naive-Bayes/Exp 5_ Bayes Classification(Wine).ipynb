**Dataset Loading and Inspection**

```python
import pandas as pd

wine_data = pd.read_csv('/content/drive/MyDrive/Datasets For ML/wine.csv')

print("Columns:", wine_data.columns)
print("\nShape:", wine_data.shape)
print("\nDuplicated:", wine_data.duplicated().sum())
print("\nNull Values:", wine_data.isnull().sum().sum())
print("\nFirst 2 rows:")
wine_data.head(2)
```

**Feature–Label Separation**

```python
X, Y = wine_data.iloc[:, 1:], wine_data.iloc[:, 0]
```

**Feature Scaling (Recommended for GaussianNB)**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Train–Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=5)
```

**Why GaussianNB for Wine?**

Features are continuous chemical measurements

Gaussian likelihood assumption holds reasonably well

MultinomialNB and BernoulliNB are not suitable

**GaussianNB Model (Wine)**

```python
from sklearn.naive_bayes import GaussianNB

gnb_wine = GaussianNB()
gnb_wine.fit(X_train, Y_train)
```

**Predictions and Probabilities**

```python
Y_pred_wine = gnb_wine.predict(X_test)
print("Predicted Labels:\n", Y_pred_wine)

print("\nPredicted Probabilities:\n", gnb_wine.predict_proba(X_test)[:10])
```

**Model Evaluation**

```python
from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(Y_test, Y_pred_wine))
print("Classification Report:\n", metrics.classification_report(Y_test, Y_pred_wine))
print("Confusion Matrix:\n", metrics.confusion_matrix(Y_test, Y_pred_wine))
```

**Exploring Model Parameters and Methods**

```python
print("Model Parameters:", gnb_wine.get_params())
print("\nClass Priors:", gnb_wine.class_prior_)
print("\nClass Means (Theta):\n", gnb_wine.theta_)
print("\nClass Variance:\n", gnb_wine.var_)
```

**Conclusion**

The Wine dataset was successfully classified using Naive Bayes, where **GaussianNB** performed best due to the continuous numerical features in the dataset. Feature scaling improved numerical stability and model performance. Overall, the model efficiently handled the dataset’s moderate dimensionality and provided accurate classification results.


**1. Bernoulli Naive Bayes**

Dataset: load_digits (binarized)

Best for: Binary / boolean features

**When to use BernoulliNB**

Features are 0/1 (present or not present)

Examples:

Word exists in document (yes/no)

Pixel is on/off

Clicked / not clicked

**1: BernoulliNB with Digits Dataset**

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import Binarizer

# Load dataset
digits = load_digits()
# print(digits.head(2)) # This line caused an AttributeError as digits is a Bunch object, not a DataFrame.
X = digits.data
y = digits.target

# Binarize features (important for BernoulliNB)
binarizer = Binarizer(threshold=8)
X_binary = binarizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_binary, y, test_size=0.2, random_state=42
)

# Model
model = BernoulliNB()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

**2. Multinomial Naive Bayes**

Dataset: 20 Newsgroups

Best for: Count-based features

**When to use MultinomialNB**

Features are counts / frequencies

Examples:

Word counts (Bag of Words)

Term frequencies

Text classification (spam, topic detection)

**2: MultinomialNB with Text Data**

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.feature_extraction.text import CountVectorizer

# Load dataset
data = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
X = data.data
y = data.target

# Convert text to word counts
vectorizer = CountVectorizer(stop_words='english')
X_counts = vectorizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_counts, y, test_size=0.2, random_state=42
)

# Model
model = MultinomialNB()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

**3. Complement Naive Bayes**

Dataset: 20 Newsgroups

Best for: Imbalanced text datasets

**When to use ComplementNB**

Text classification

Classes are imbalanced

Improves over MultinomialNB for skewed data

Often used in real-world NLP problems

**3: ComplementNB with Text Data**

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.feature_extraction.text import CountVectorizer

# Load dataset
data = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
X = data.data
y = data.target

# Vectorize text
vectorizer = CountVectorizer(stop_words='english')
X_counts = vectorizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_counts, y, test_size=0.2, random_state=42
)

# Model
model = ComplementNB()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
