**SciPy Library**

SciPy is a Python library used for scientific and mathematical computations.
It is built on top of NumPy and provides advanced functions for optimization, statistics, algebra, integration, and signal processing.
SciPy is commonly used in Machine Learning for optimization, statistical analysis, and numerical methods.

**SciPy Constants Module:** Provides commonly used mathematical and physical constants.

**Importing SciPy Modules**

```python
from scipy import constants
```

**Viewing Available Constants:** Displays all constants available in the SciPy constants module.

```python
dir(constants)
```

**Euler’s Number (e):** Represents the base of natural logarithms.

```python
constants.e
```

**Zetta Prefix:** Represents the SI prefix value for zetta (10²¹).

```python
constants.zetta
```

**SciPy Optimize Module:** Used to find roots, minimums, and maximums of mathematical functions.
Very important in Machine Learning for cost function optimization.

```python
from scipy import optimize
```

**Defining an Equation:** Defines a mathematical function to be optimized.

```python
def equation(x):
    return x**4 + 3*x**3 + 3*x**2 - 2*x + 1
```

**Finding Roots of an Equation**

**Root Finding:** Finds the value of x where the equation becomes zero.

```python
r = optimize.root(equation, 0)
r
```

**Extracting Root Value:** Returns the numerical root value.

```python
r.x
```

**Finding Minimum Value of a Function**

**Function Minimization:** Finds the value of x where the function has minimum value.

```python
m = optimize.minimize(equation, 0)
m
```

**Extracting Minimum Point:** Returns the x-value at which the function is minimum.

```python
m.x
```

**Why Optimization Matters in ML**

*   Used in training models

*   Helps minimize loss or cost functions

*   Improves accuracy of predictions

*   Core concept in regression and neural networks

**SciPy Statistics Module (Common in ML)**

The scipy.stats module is used for probability distributions and advanced statistical analysis.
Basic operations like mean and median are handled by NumPy for better performance, while SciPy is used for statistical tests and analysis.

```python
from scipy import stats
```

**Sample Data**

```python
data1 = [10, 20, 30, 40, 50]
data2 = [15, 25, 35, 45, 55]
```

**Mode Calculation**

**Finding Mode:** Returns the most frequently occurring value in the dataset.
Used when identifying the most common value.

```python
stats.mode(data1)
```

**Descriptive Statistics**

**Describe Method:** Provides summary statistics such as count, mean, variance, min, and max.

```python
stats.describe(data1)
```

**Skewness:** Measures the asymmetry of the data distribution.

```python
stats.skew(data1)
```

**Kurtosis:** Measures the peakedness or flatness of the data distribution.

```python
stats.kurtosis(data1)
```

**Independent T-Test**

**T-Test:** Compares the means of two independent datasets.
Used to check whether the difference between two datasets is statistically significant.

```python
stats.ttest_ind(data1, data2)
```

**Pearson Correlation:** Measures the strength and direction of linear relationship between two datasets.

```python
stats.pearsonr(data1, data2)
```

**Distance Calculation (ML Use Case)**

**Euclidean Distance:** Used in KNN and clustering algorithms.

```python
from scipy.spatial import distance
```

```python
a = [1, 2]
b = [4, 6]

distance.euclidean(a, b)
```
