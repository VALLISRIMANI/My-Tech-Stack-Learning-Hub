**1. Bernoulli Naive Bayes**

Dataset: load_digits (binarized)

Best for: Binary / boolean features

**When to use BernoulliNB**

Features are 0/1 (present or not present)

Examples:

Word exists in document (yes/no)

Pixel is on/off

Clicked / not clicked

**1: BernoulliNB with Digits Dataset**

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import Binarizer

# Load dataset
digits = load_digits()
# print(digits.head(2)) # This line caused an AttributeError as digits is a Bunch object, not a DataFrame.
X = digits.data
y = digits.target

# Binarize features (important for BernoulliNB)
binarizer = Binarizer(threshold=8)
X_binary = binarizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_binary, y, test_size=0.2, random_state=42
)

# Model
model = BernoulliNB()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

**2. Multinomial Naive Bayes**

Dataset: 20 Newsgroups

Best for: Count-based features

**When to use MultinomialNB**

Features are counts / frequencies

Examples:

Word counts (Bag of Words)

Term frequencies

Text classification (spam, topic detection)

**2: MultinomialNB with Text Data**

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.feature_extraction.text import CountVectorizer

# Load dataset
data = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
X = data.data
y = data.target

# Convert text to word counts
vectorizer = CountVectorizer(stop_words='english')
X_counts = vectorizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_counts, y, test_size=0.2, random_state=42
)

# Model
model = MultinomialNB()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

**3. Complement Naive Bayes**

Dataset: 20 Newsgroups

Best for: Imbalanced text datasets

**When to use ComplementNB**

Text classification

Classes are imbalanced

Improves over MultinomialNB for skewed data

Often used in real-world NLP problems

**3: ComplementNB with Text Data**

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.feature_extraction.text import CountVectorizer

# Load dataset
data = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
X = data.data
y = data.target

# Vectorize text
vectorizer = CountVectorizer(stop_words='english')
X_counts = vectorizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_counts, y, test_size=0.2, random_state=42
)

# Model
model = ComplementNB()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
